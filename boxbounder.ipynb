```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
from IPython.display import clear_output
```

```python
# list of all dataset "scenarios"
names = ["basketball", "boxes", "football", "juggle", "softball", "tennis"]
namesidx = 0
camidx = 0
running = True


while running:
    inpath = f'./data/{names[namesidx]}/ims/{camidx}/' # input frames
    outpath = f'./annotations/{names[namesidx]}/{camidx}.txt' # output txts

    frame_names = [
        p for p in os.listdir(inpath)
        if os.path.splitext(p)[-1] in [".jpg", ".jpeg", ".JPG", ".JPEG"]
    ]
    frame_names.sort(key=lambda p: int(os.path.splitext(p)[0]))

    boxes = []

    try:
        textf = open(outpath, "r")
        boxes = textf.readlines()
        boxes = [[int(chara) for chara in line.split(", ")] for line in boxes]
        textf.close()
    except OSError:
        print(f"failed to read {names[namesidx]}/{camidx}/{outpath} somehow, using fresh file")
        boxes = []
    except ValueError:
        print(f"{names[namesidx]}/{camidx}/{outpath} has bad format, using fresh file")
        boxes = []

    # the main seg loop
    '''
    modes:
    m   menu; here by default
    s   segment; add a bbox

    menu keybinds:
    s or space
        go into segment mode
    c   exit segment mode
    w   write, i.e. save
    f   delete all bboxes
    e   continue to next
    q   go to last
    g   switch video by input
    h   switch cam by input
    1   stop
    '''
    inseg = True
    mode = "m" 
    while inseg:
        clear_output()
        if mode == "m":
            img = cv2.imread(os.path.join(inpath, frame_names[0]))
            for box in boxes:
                img = cv2.rectangle(img, box[:2], box[2:], (0, 255, 0), 2)
            cv2.imshow("MENU", img)
            print(f"{names[namesidx]}/{camidx} MENU")
            ret = cv2.waitKey(0)

            if ret == ord("s") or ret == ord(" "):
                mode = "s"

            elif ret == ord("w"):
                textf = open(outpath, "w")
                for line in boxes:
                    textf.write(str(line).strip("[]")+"\n")
                textf.close()

            elif ret == ord("f"):
                boxes = []

            elif ret == ord("e"):
                inseg = False
                if camidx > len(os.listdir(f'./data/{names[namesidx]}/ims/')) - 2:
                    if namesidx < len(names)-1:
                        namesidx += 1
                        camidx = 0
                else:
                    camidx += 1

            elif ret == ord("q"):
                inseg = False
                if camidx <= 0:
                    if namesidx > 0:
                        namesidx -= 1
                        camidx = len(os.listdir(f'./data/{names[namesidx]}/ims/')) - 1
                else:
                    camidx -= 1

            elif ret == ord("g"):
                inseg = False
                cv2.destroyAllWindows()
                namesidx = int(input("Input desired video by idx: "))

            elif ret == ord("h"):
                inseg = False
                cv2.destroyAllWindows()
                camidx = int(input("Input desired cam by idx: "))

            elif ret == ord("1"):
                cv2.destroyAllWindows()
                inseg = False
                running = False
        
        
        elif mode == "s":
            img = cv2.imread(os.path.join(inpath, frame_names[0]))
            for box in boxes:
                img = cv2.rectangle(img, box[:2], box[2:], (0, 255, 0), 2)
            resultbox = cv2.selectROI("SEGMENT", img)
            if resultbox == (0, 0, 0, 0):
                mode = "m"
                continue
            print(f"{names[namesidx]}/{camidx} SEGMENT")
            resultbox = [resultbox[0], resultbox[1], resultbox[0] + resultbox[2], resultbox[1] + resultbox[3]]
            boxes.append(resultbox)
            textf = open(outpath, "w")
            for line in boxes:
                textf.write(str(line).strip("[]")+"\n")
            textf.close()



```

```python
import os

names = ["basketball", "boxes", "football", "juggle", "softball", "tennis"]

for name in names:
    try:
        os.makedirs(f"./annotations/{name}")
    except FileExistsError:
        print("dir", f"./annotations/{name}", "exists")

    for cam in range(31):
        try:
            os.makedirs(f"./output/{name}/{cam}/")
        except FileExistsError:
            print("dir", f"./output/{name}/{cam}/", "exists")
```
